## Preprocessing Data

```python
# Make Datetime values
data['Sales date']=data['Sales date'].apply(pd.to_datetime)
```

```python
#Year value was incorrect cause no machine was made in year 1000. Replace those by nans
data['Year Made'] = data['Year Made'].replace(1000, np.nan)
```

```python
data.sample(5)
```

## Basic Overview

Try to find what the relation of the data to the to be predicted value is. Start with the data columns that are (almost) complete and do contain a limited number of unique values.

```python
sns.distplot(data['Sales Price'])
#To start with, just a Distplot of sales price to see what we are dealing with
```

```python
#To check the relation with the Year Made we use a regplot 
sns.regplot(x='Year Made', y='Sales Price', data=data)
plt.xlabel('Year Made')
plt.ylabel('Sales Price')
# In general Year Made has a positive correlation with the Sales Price, but the data is widely scattered.
```

```python
#To check the relation with the Year Sale we use a regplot 
data['Year Sale']=data['Sales date'].dt.year
sns.regplot(x='Year Sale', y='Sales Price', data=data)
plt.xlabel('Sales date')
plt.ylabel('Sales Price')
# In general Year Sale has a weak correlation with sales price. We can dig further to month, day of week, week but I cannot imagine
# this has a vast impact on the sales price.
```

```python
#To check the relation with the machine size we use a boxplot 
sns.boxplot(x='Machine Size', y='Sales Price', data=data)
plt.xlabel('Machine Size')
plt.ylabel('Sales Price')
data.groupby('Machine Size')['Sales Price'].mean()
# Large Difference between price values indicate Machine Size seems very important feature in determining the price
```

```python
#To check the relation with the model ID we use a scatterplot 
sns.scatterplot(x='Model ID', y='Sales Price', data=data)
plt.xlabel('Model ID')
plt.ylabel('Sales Price')
data.groupby('Model ID')['Sales Price'].mean()
#Even tough model has same ID, price range can very a lot (see plot)
```

```python
#To check the relation with the datasource we use a boxplot 
sns.boxplot(x='datasource', y='Sales Price', data=data)
plt.xlabel('datasource')
plt.ylabel('Sales Price')
data.groupby('datasource')['Sales Price'].mean()
# datasource seems as a relevant feature in determining the price, 173 is an outlier?
```

```python
#To check the relation with the Model Description we use a a scatterplot 
#sns.scatterplot(x='Model Description', y='Sales Price', data=data)
#plt.xlabel('Model Description')
#plt.ylabel('Sales Price')
data.groupby('Model Description')['Sales Price'].mean().sort_values().reset_index()
#The model is relevant in determining the price, but there are many unique models'
```

```python
#To check the relation with the Auctioneer ID we use a a scatterplot, seems like there are outliers at ID 100 
sns.scatterplot(x='Auctioneer ID', y='Sales Price', data=data)
plt.xlabel('Auctioneer ID')
plt.ylabel('Sales Price')
data.groupby('Auctioneer ID')['Sales Price'].mean()
```

```python
#To check the relation with the Usage Band we use a a boxplot 
sns.boxplot(x='Usage Band', y='Sales Price', data=data)
plt.xlabel('Usage Band')
plt.ylabel('Sales Price')
data.groupby('Usage Band')['Sales Price'].mean()
# Usage Band seems an important feature in determining the price, but the data is incomplete
```

```python
#To check the relation with the Machine Size we use a boxplot 
sns.boxplot(x='Machine Size', y='Sales Price', data=data)
plt.xlabel('Machine Size')
plt.ylabel('Sales Price')
data.groupby('Machine Size')['Sales Price'].mean()
# Enclosure seems an important feature in determining the price but data is incompete
```
